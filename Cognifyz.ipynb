```python

import pandas as pd

# Load the dataset
file_path = '/content/sample_data/Dataset .csv'
df = pd.read_csv(file_path)
print(df.head(10))
```

```python
missing_values=df.isnull().sum()
print("The missing values in this dataset are : \n",missing_values)
```

```python
#Task1:Top Cuisines

cuisine_column = 'Cuisines'
if cuisine_column not in df.columns:
    print(f"Please replace 'Cuisine_Column' with the actual column name for cuisines in your dataset.")
else:
    # Calculate the top 3 most common cuisines
    top_cuisines = df[cuisine_column].value_counts().head(3)

    # Calculate the percentage of restaurants serving each top cuisine
    total_restaurants = len(df)
    top_cuisines_percentage = (top_cuisines / total_restaurants) * 100

    # Display the results
    print("Top 3 Cuisines:")
    print(top_cuisines)
    print("\nPercentage of Restaurants Serving Each Top Cuisine:")
    print(top_cuisines_percentage)
```

```python
#Task2:City analysis

city_column = 'City'
rating_column = 'Aggregate rating'

if city_column not in df.columns or rating_column not in df.columns:
    print(f"Please replace 'City_Column' and 'Rating_Column' with the actual column names in your dataset.")
else:
    # 1. Identify the city with the highest number of restaurants
    city_restaurant_counts = df[city_column].value_counts()
    city_with_most_restaurants = city_restaurant_counts.idxmax()
    most_restaurants_count = city_restaurant_counts.max()

    # 2. Calculate the average rating for restaurants in each city
    avg_ratings_by_city = df.groupby(city_column)[rating_column].mean()

    # 3. Determine the city with the highest average rating
    city_with_highest_avg_rating = avg_ratings_by_city.idxmax()
    highest_avg_rating = avg_ratings_by_city.max()

    # Display the results
    print(f"The city with the highest number of restaurants is '{city_with_most_restaurants}' with {most_restaurants_count} restaurants.")
    print("\nAverage Ratings by City:")
    print(avg_ratings_by_city)
    print(f"\nThe city with the highest average rating is '{city_with_highest_avg_rating}' with an average rating of {highest_avg_rating:.2f}.")
```

```python
#Task3:Prize range distributions

import matplotlib.pyplot as plt

price_range_column = 'Price range'

if price_range_column not in df.columns:
    print(f"Please replace 'Price_Range_Column' with the actual column name for price ranges in your dataset.")
else:
    # 1. Calculate the distribution of price ranges
    price_range_counts = df[price_range_column].value_counts()

    # 2. Calculate the percentage of restaurants for each price range
    total_restaurants = len(df)
    price_range_percentages = (price_range_counts / total_restaurants) * 100

    # Display the distribution and percentages
    print("Price Range Distribution:")
    print(price_range_counts)
    print("\nPrice Range Percentages:")
    print(price_range_percentages)

    # 3. Create a bar chart to visualize the distribution of price ranges
    plt.figure(figsize=(10, 6))
    price_range_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Distribution of Price Ranges Among Restaurants', fontsize=16)
    plt.xlabel('Price Range', fontsize=14)
    plt.ylabel('Number of Restaurants', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
#Task1:restaurant ratings
rating_column = 'Aggregate rating'
votes_column = 'Votes'

if rating_column not in df.columns or votes_column not in df.columns:
    print(f"Please replace 'Rating_Column' and 'Votes_Column' with the actual column names in your dataset.")
else:
    # 1. Analyze the distribution of ratings and find the most common rating range
    rating_bins = [0, 1, 2, 3, 4, 5]  # Define bins for rating ranges
    df['Rating_Range'] = pd.cut(df[rating_column], bins=rating_bins)
    rating_distribution = df['Rating_Range'].value_counts().sort_index()

    # Find the most common rating range
    most_common_rating_range = rating_distribution.idxmax()
    most_common_count = rating_distribution.max()

    # Display the rating distribution and most common rating range
    print("Rating Distribution:")
    print(rating_distribution)
    print(f"\nMost Common Rating Range: {most_common_rating_range} with {most_common_count} restaurants.")

    # Plot the rating distribution
    plt.figure(figsize=(8, 6))
    rating_distribution.plot(kind='bar', color='lightgreen', edgecolor='black')
    plt.title('Distribution of Ratings', fontsize=16)
    plt.xlabel('Rating Range', fontsize=14)
    plt.ylabel('Number of Restaurants', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # 2. Calculate the average number of votes received by restaurants
    avg_votes = df[votes_column].mean()
    print(f"\nThe average number of votes received by restaurants is {avg_votes:.2f}.")
```

```python
#Task2:cuisines combination

cuisine_column = 'Cuisines'
rating_column = 'Aggregate rating'

if cuisine_column not in df.columns or rating_column not in df.columns:
    print(f"Please replace 'Cuisine_Column' and 'Rating_Column' with the actual column names in your dataset.")
else:
    df[cuisine_column] = df[cuisine_column].fillna('')
    df[cuisine_column] = df[cuisine_column].astype(str)

    # 1. Identify the most common cuisine combinations
    df['Cuisine_List'] = df[cuisine_column].str.split(', ')
    df['Cuisine_Combination'] = df['Cuisine_List'].apply(lambda x: ', '.join(sorted(x)) if isinstance(x, list) else '')
    common_combinations = df['Cuisine_Combination'].value_counts()

    # Display the top 10 most common cuisine combinations
    print("Top 10 Most Common Cuisine Combinations:")
    print(common_combinations.head(10))

    # 2. Determine if certain cuisine combinations have higher ratings
    avg_rating_by_combination = (
        df.groupby('Cuisine_Combination')[rating_column].mean().sort_values(ascending=False)
    )

    # Display the top 10 cuisine combinations with the highest ratings
    print("\nTop 10 Cuisine Combinations with the Highest Average Ratings:")
    print(avg_rating_by_combination.head(10))
```

```python


import folium
from folium.plugins import MarkerCluster


latitude_column = 'Latitude'
longitude_column = 'Longitude'
restaurant_name_column = 'Restaurant Name'

if latitude_column not in df.columns or longitude_column not in df.columns:
    print(f"Please replace 'Latitude_Column' and 'Longitude_Column' with the actual column names in your dataset.")
else:

    df = df.dropna(subset=[latitude_column, longitude_column])

    # Initialize a folium map
    avg_lat = df[latitude_column].mean()
    avg_lon = df[longitude_column].mean()
    restaurant_map = folium.Map(location=[avg_lat, avg_lon], zoom_start=12)

    # Add markers for each restaurant
    marker_cluster = MarkerCluster().add_to(restaurant_map)

    for _, row in df.iterrows():
        lat = row[latitude_column]
        lon = row[longitude_column]
        name = row.get(restaurant_name_column, "Unknown Restaurant")  # Fallback if name column is missing
        folium.Marker(
            location=[lat, lon],
            popup=f"<b>{name}</b><br>Lat: {lat}, Lon: {lon}",
            icon=folium.Icon(color='blue', icon='info-sign')
        ).add_to(marker_cluster)

    # Save the map as an HTML file
    map_file_path = "restaurant_locations_map.html"
    restaurant_map.save(map_file_path)

    #download the html file
    from google.colab import files
    files.download("restaurant_locations_map.html")



    print(f"Map has been saved as {map_file_path}. You can open it in a browser to view.")


```

```python
import pandas as pd
import geopandas as gpd
import folium
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Drop rows with missing coordinates
df = df.dropna(subset=['Latitude','Longitude'])

# Detect clusters using DBSCAN
coords = df[['Latitude', 'Longitude']].to_numpy()
dbscan = DBSCAN(eps=0.01, min_samples=5, metric='haversine')  # Adjust 'eps' as needed
dbscan.fit(coords)

# Add cluster labels to the dataset
df['cluster'] = dbscan.labels_

# Plot clusters using Matplotlib
plt.figure(figsize=(10, 8))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['cluster'], cmap='rainbow', alpha=0.6)
plt.colorbar(scatter, label='Cluster Label')
plt.title('Restaurant Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python


data = df.dropna(subset=['Restaurant Name', 'Latitude', 'Longitude'])
chains = data.groupby('Restaurant Name').size().reset_index(name='location_count')
restaurant_chains = chains[chains['location_count'] > 1]

# Output the result
if restaurant_chains.empty:
    print("No restaurant chains found in the dataset.")
else:
    print(f"Restaurant chains found ({len(restaurant_chains)}):")
    print(restaurant_chains)
```

```python

required_columns = ['Restaurant Name', 'Aggregate rating', 'Votes']
missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    print(f"\nMissing required columns in the dataset: {missing_columns}")
else:
    # Identify restaurant chains
    chain_counts = df['Restaurant Name'].value_counts()
    restaurant_chains = chain_counts[chain_counts > 1]

    print("\nIdentified Restaurant Chains:")
    print(restaurant_chains)

    # Filter dataset to include only restaurant chains
    chain_data = df[df['Restaurant Name'].isin(restaurant_chains.index)]

    # Analyze ratings for each chain
    ratings_analysis = chain_data.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False)
    print("\nAverage Ratings of Restaurant Chains:")
    print(ratings_analysis)

    # Analyze popularity for each chain
    popularity_analysis = chain_data.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending=False)
    print("\nPopularity of Restaurant Chains:")
    print(popularity_analysis)

    # Combine ratings and popularity into a single summary
    analysis_summary = pd.DataFrame({
        'Average Rating': ratings_analysis,
        'Total Popularity': popularity_analysis
    }).sort_values(by='Total Popularity', ascending=False)

    print("\nSummary Analysis of Restaurant Chains:")
    print(analysis_summary)

    # save the analysis results to a CSV file
    output_path = "restaurant_chain_analysis.csv"
    analysis_summary.to_csv(output_path)
    print(f"\nAnalysis results saved to {output_path}")

    #download csv
    from google.colab import files
    files.download("restaurant_chain_analysis.csv")

```
